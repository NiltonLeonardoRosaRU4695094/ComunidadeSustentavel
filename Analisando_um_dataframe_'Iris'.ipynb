```
import matplotlib.pyplot as plt

#Calcular a frequência absoluta das especies
species_counts = data['species'].value_counts()

#criar um grafico de barras
fig, ax = plt.subplots(figsize=(8, 6))
species_counts.plot(kind='bar', ax=ax, color='skyblue')

#adicionar rotulos e titulo
ax.set_title('Total de espécies')
ax.set_xlabel('Espécie')
ax.set_ylabel('Total')

#Manter os rotulos do eixo x na horizontal
plt.xticks(rotation=0)

#Exibir grafico
plt.tight_layout()
plt.show()
```

```
#Calcular a frequência absoluta das especies
species_counts = data['species'].value_counts()

#criar um grafico de barras
fig, ax = plt.subplots(figsize=(8, 6))
species_counts.plot(kind='barh', ax=ax, color='lightgreen')

#adicionar rotulos e titulo
ax.set_title('Total de espécies')
ax.set_ylabel('Total')
ax.set_xlabel('Espécie')

plt.xticks(rotation=360)

#Exibir grafico
plt.tight_layout()
plt.show()
```

```
import seaborn as sns

# Selecionar apenas as colunas numéricas para o boxplot
numerical_data = data.select_dtypes(include=['number'])

# Criar boxplots para todas as colunas numéricas
plt.figure(figsize=(10, 6))
sns.boxplot(data=numerical_data)
plt.title('Boxplots de Todas as Colunas Numéricas')
plt.ylabel('Valor')
# Rotacionar os rótulos do eixo x para melhor legibilidade
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```
# Agrupar os dados por espécie e calcular a média
mean_by_species = data.groupby('species').mean()

print("Média agrupada por espécie:")
display(mean_by_species)

# Agrupar os dados por espécie e calcular a mediana
median_by_species = data.groupby('species').median()

print("\nMediana agrupada por espécie:")
display(median_by_species)
```

```
# Agrupar os dados por espécie e calcular o desvio padrão
std_by_species = data.groupby('species').std()

print("Desvio padrão agrupado por espécie:")
display(std_by_species)

# Agrupar os dados por espécie e calcular a variância
var_by_species = data.groupby('species').var()

print("\nVariância agrupada por espécie:")
display(var_by_species)
```

As amostras apresentam uma frequência equilibrada, com 50 especies em cada um dos 3 lote. Com uma frequência relatia em 0,33, apresentando uma pequena diferença no desvio padrão e na variança. As diferenças apresentadas pela estatistica e no Boxplots mostram suas caracteristicas das petalas são uteis para distinguir as espécies.
